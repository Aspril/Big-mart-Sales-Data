### FEATURE ENGINEERING

AS ITEM FAT CONTENT COLUMNS HAS ATTRIBUTES LIKE (LF,LOW FAT,low fat) and (Regular,reg) WE WILL COBMINE THEM.

```python
data["Item_Fat_Content"].replace(["LF","low fat"],"Low Fat",inplace=True)

data["Item_Fat_Content"].replace("reg","Regular",inplace=True)
```

ITEM VISIBILITY COLUMN HAS SOME VALUES AS 0 (VISIBILITY MAY BE POOR BUT NOT 0 IF THE PRODUCT IS IN THE STORE)
THEREFORE I WILL REPLACE 0 WITH MEAN.

```python
data["Item_Visibility"]=(data["Item_Visibility"]==0).apply(lambda x:data["Item_Visibility"].mean())
```

```python
Item_visibility_avg=data.pivot_table(values="Item_Visibility",index="Item_Identifier")
```

```python
#data['Item_Visibility_MeanRatio'] = data.apply(lambda x: x['Item_Visibility']/Item_visibility_avg[x['Item_Identifier']], axis=1)
```

```python
data.pivot_table(values=["Item_Outlet_Sales"],index="Outlet_Type")
```

AS MEAN VALUE OF ITEM OUTLET SALES FOR EACH OUTLET TYPE IS DIFFERENT, I WILL NOT COMBINE THE OUTLET TYPES

-------------------------------------------------------------------------------------------------------------------------

ITEM TYPE HAS 16 CATEGORIES, LOOKING CLOSELY WITH ITEM IDENTIFIER ITEMS WHICH ARE DRINKS HAVE 'DR' IN ITEM IDENTIFIER,SIMILARY NON CONSUMABLE:"NC", FOOD: "FD". SO WE CAN COMBINE 16 CATEGORIES IN 3 MAIN CATEGORIES. 

```python
data["Item_type_combined"]=data["Item_Identifier"].apply(lambda x: x[0:2])

data["Item_type_combined"]=data["Item_type_combined"].replace({"FD":"Food","NC":"Non consumable","DR":"Drinks"})
```

FROM OUTLET ESTABLISHMENT YEAR I CAN FIND SINCE HOW MANY YEARS THE OUTLET HAS BEEN RUNNING

```python
data['No_of_years']=2018-data["Outlet_Establishment_Year"]
```

```python
data.loc[data["Item_type_combined"]=="Non consumable","Item_Fat_Content"]="Non Edible"
```

### ONE HOT ENCODING

```python
categorical_columns=[x for x in data.columns if data.dtypes[x]=="object"]

#categorical_columns.remove("source",'Item_Identifier','Outlet_Identifier')
categorical_columns=[x for x in categorical_columns if x not in ["source",'Item_Identifier','Outlet_Identifier','Item_Type']]

categorical_columns
```

### BUILDING A MODEL

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error

test=data[data["source"]=="test"]

train=data[data["source"]=="train"]


train=train.drop(columns=["Item_Identifier","Outlet_Identifier","source","Item_Type"])
test=test.drop(columns=["Item_Identifier","Outlet_Identifier","source","Item_Type"])

y=train.Item_Outlet_Sales.values
X=train.drop(columns=["Item_Outlet_Sales"],axis=0).values

X_train,X_cv,y_train,y_cv=train_test_split(X,y,test_size=0.3)

test=test.drop(columns=["Item_Outlet_Sales"],axis=0).values

lin_model=LinearRegression()

lin_model.fit(X_train,y_train)

y_pred=lin_model.predict(X_cv)

print("MSE: %.4f" % np.sqrt(mean_squared_error(y_cv,y_pred)))

print('Variance score: %.4f' % r2_score(y_cv, y_pred))
```
